```python
import torch
from ultralytics import YOLO


# config (example)
DATA_YAML = "data.yaml"
MODEL = "yolov8s.pt"
EPOCHS = 10
IMG_SIZE = 640
BATCH_SIZE = 16
LR0 = 0.01
PROJECT_DIR = "yolo_runs"
RUN_NAME = "colab_safe_run"

# choose device safely
if torch.cuda.is_available():
    device = 0  # use GPU 0
    print("Using CUDA GPU (device=0).")
else:
    device = "cpu"
    print("CUDA not available ‚Äî using CPU. To enable GPU, change runtime to GPU and install CUDA-enabled torch.")

# adjust batch for CPU if required
if device == "cpu":
    # reduce batch size to avoid OOM / slow performance on CPU
    if BATCH_SIZE is None or BATCH_SIZE > 8:
        BATCH_SIZE = 4
        print("Running on CPU: set BATCH_SIZE=4 to avoid OOM / extremely slow training.")

model = YOLO(MODEL)
print("Starting training on device:", device)
model.train(
    data=DATA_YAML,
    epochs=EPOCHS,
    imgsz=IMG_SIZE,
    batch=BATCH_SIZE,
    lr0=LR0,
    project=PROJECT_DIR,
    name=RUN_NAME,
    device=device,
    save_period=10,
    pretrained=True
)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Path to your results.csv (update if needed)
csv_path = "yolo_runs/colab_safe_run4/results.csv"

# Load CSV
df = pd.read_csv(csv_path)

print("Columns in results.csv:", list(df.columns))
print(df.head())

# Plot loss curves
plt.figure(figsize=(10,5))
if 'train/box_loss' in df and 'train/cls_loss' in df and 'train/dfl_loss' in df:
    plt.plot(df['epoch'], df['train/box_loss'], label='box_loss')
    plt.plot(df['epoch'], df['train/cls_loss'], label='cls_loss')
    plt.plot(df['epoch'], df['train/dfl_loss'], label='dfl_loss')
else:
    print("Loss columns not found in results.csv")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Losses")
plt.legend()
plt.grid(True)
plt.show()

# Plot precision, recall, mAP
plt.figure(figsize=(10,5))
for metric in ['metrics/precision(B)', 'metrics/recall(B)', 'metrics/mAP50(B)', 'metrics/mAP50-95(B)']:
    if metric in df:
        plt.plot(df['epoch'], df[metric], label=metric)
plt.xlabel("Epoch")
plt.ylabel("Metric Value")
plt.title("Validation Metrics")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import os
from ultralytics import YOLO
from IPython.display import display
from PIL import Image
import glob

# Config - change these paths as needed
# model_path can be "runs/train/your_run_name/weights/best.pt" or "yolov8s.pt" or "runs/train/your_run_name/weights/last.pt"
# If you trained in the previous run, we'll try to pick best.pt automatically.
def find_ckpt(run_dir=None):
    if run_dir is None:
        # try default latest run
        run_dir = sorted(glob.glob("yolo_runs/colab_safe_run4"), key=os.path.getmtime)[-1]
    weights_dir = os.path.join(run_dir, "weights")
    candidates = [
        os.path.join(weights_dir, "best.pt"),
        os.path.join(weights_dir, "last.pt"),
        os.path.join(run_dir, "best.pt"),
        os.path.join(run_dir, "last.pt"),
    ]
    for c in candidates:
        if os.path.exists(c):
            return c
    return None

# Set your image source: single image path or directory
IMAGE_OR_DIR = "test/images"  # change to folder with images or single image path
SAVE_DIR = "predictions"
os.makedirs(SAVE_DIR, exist_ok=True)

# Determine model path
ckpt = find_ckpt()  # tries latest run
if ckpt:
    print("Found checkpoint:", ckpt)
    model_path = ckpt
else:
    # fallback to a pretrained model from ultralytics package
    print("No checkpoint found in runs/train. Falling back to pretrained 'yolov8s.pt'.")
    model_path = "yolov8s.pt"

# Load model
model = YOLO(model_path)

# Get image list
img_list = []
if os.path.isdir(IMAGE_OR_DIR):
    for ext in ("*.jpg", "*.jpeg", "*.png", "*.bmp"):
        img_list.extend(glob.glob(os.path.join(IMAGE_OR_DIR, ext)))
else:
    img_list = [IMAGE_OR_DIR]

if not img_list:
    raise FileNotFoundError(f"No images found in {IMAGE_OR_DIR}. Put images there or change IMAGE_OR_DIR.")

print(f"Running inference on {len(img_list)} images...")

# Run inference and save visualized predictions
for img_path in img_list:
    # predict
    results = model.predict(source=img_path, save=False, verbose=False, imgsz=640)  # returns ultralytics Results
    # results is list-like; handle first result
    res = results[0]
    # save annotated image to SAVE_DIR (Ultralytics provides a .plot() or .save() method)
    # Use res.plot() to get numpy image with boxes drawn:
    annotated = res.plot()  # numpy array HxWxC
    # Save annotated image
    out_name = os.path.join(SAVE_DIR, os.path.basename(img_path))
    Image.fromarray(annotated).save(out_name)
    display(Image.fromarray(annotated))
    print("Saved:", out_name)
```
